### Basic Hypothesis Testing in Python

This is the same sales data we have seen in Excel

```python
# upload the csv file and create a dataframe

import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv('pairedData.csv')
```

```python
df.head()
```

Perform some basic EDA - we will do more next week


```python
print('Mean Access_2018: ', df['Access_2018'].mean())
print('Mean Access_2019: ', df['Access_2019'].mean())
```

This is paired data so we are more interested in paired differences

```python
df['diff'] = df['Access_2019'] - df['Access_2018']
```

```python
df.head(20)
```

```python
print('Mean difference is: ', df['diff'].mean())
```

### Hypothesis Test (paired t test)
### Null Hypothesis: There is no difference between Access 2018 and Access 2019
### Alternative Hypothesis: There is a difference (2 sided)

```python
stats.ttest_rel(df['Access_2018'], df['Access_2019'])
```

```python
stats.ttest_rel(df['Access_2018'], df['Access_2019'], alternative='two-sided')
```

```python
help(stats.ttest_rel)
```

If this was not paired data and we wanted to do a 2 sample t test to compare means.

```python
stats.ttest_ind(df['Access_2018'], df['Access_2019'])
```
